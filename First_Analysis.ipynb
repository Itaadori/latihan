```python
import numpy as np
import pandas as pd

# seberapa banyak orang yang bisa buka rekening X >= 19 tahun
# dan dengan isi rekening lebih dari 0 (setelah data dibersihkan)
# pada tahun 2025

url = "https://raw.githubusercontent.com/Itaadori/latihan/refs/heads/main/test.csv"
df = pd.read_csv(url, sep=';') # menjadikan data lebih rapih agar mudah dibaca
df['balance'] = pd.to_numeric(df['balance'], errors='coerce') # mengubah menjadi numerik supaya bisa diganti minimal ke 0
df.loc[df['balance'] < 0, 'balance'] = 0 # semua nilai minus menjadi 0

df.loc[(df['age'] >= 19) & (df['age'] <= 81)] # mengetahui jumlah potensi pembeli berumur lebih dari sama dengan 19 tahun
df.loc[(df['balance'] > 0)] # mengetahui jumlah potensi pembeli dengan balance lebih dari 0
```
